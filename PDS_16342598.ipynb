A) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.

B) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
#importing required libraries for preprocessing
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from datetime import datetime
```

```python
#Reading the dataset from the file train.csv
d_frame = pd.read_csv("train.csv")
```

```python
d_frame.head()
```

```python
#checking the total number of null values in each column
null_values = d_frame.isnull().sum()
print(null_values)
```

```python
#The New_Price column is dropped as it is having lot of null values
d_frame = d_frame.drop('New_Price', axis=1)
```

```python
d_frame.head()
```

```python
d_frame['Mileage']=d_frame['Mileage'].str.split(' ').str[0]
d_frame['Engine']=d_frame['Engine'].str.split(' ').str[0]
d_frame['Power']=d_frame['Power'].str.split(' ').str[0]
```

```python
d_frame['Mileage'] = d_frame['Mileage']. astype(float)
d_frame['Engine'] = d_frame['Engine']. astype(float)
d_frame['Power'] = d_frame['Power']. astype(float)
```

```python
#replacing the mileage , engine , power , price columns with mean

mean_columns = ['Mileage', 'Engine', 'Power','Price']

# replacing the seats column with mode

mode_columns=['Seats']

mean_imputer = SimpleImputer(strategy='mean')

mode_imputer=SimpleImputer(strategy='most_frequent')

d_frame[mean_columns] = mean_imputer.fit_transform(d_frame[mean_columns])

d_frame[mode_columns]=mode_imputer.fit_transform(d_frame[mode_columns])
```

```python
#making sure the null values are removed
null_values = d_frame.isnull().sum()
print(null_values)
```

C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value

```python
d_frame = pd.get_dummies(d_frame, columns=['Fuel_Type', 'Transmission'], prefix=['Fuel','Transmission'])
```

```python
d_frame
```

```python
col = ['Fuel_Diesel','Fuel_Electric','Fuel_Petrol','Transmission_Automatic','Transmission_Manual']
#converting spercified columns to integers
d_frame[col] = d_frame[col].astype(int)
```
